## Merging Dataframes in pandas
df1.append(df2) -> vertical - by rows 

df1.join(df2) -> join on indexes

#################################################

use these : 

pd.concat([df1,df2]) -> vertical/horizontal - inner/outer join 


pd.merge([df1,df2]) -> many joins 

```python
import pandas as pd
```

```python
#focus on merge dataframe : 
```

```python
train=pd.read_csv("http://bit.ly/kaggletrain")
```

```python
train.head(2)
```

```python
#Create 2 dataset , and PassengerId is common on both : 
train1=train.loc[:,('PassengerId','Pclass','Name')].copy()
train2=train.loc[:,('PassengerId','Sex','Age')].copy()
print(train1.head(2))
print(train2.head(2))
```

```python
#now merge both the dataset using merge : 

merged_data=pd.merge(train1,train2)
print(merged_data.head(2))
```

```python
#now set index as PassengerId
train1.set_index('PassengerId',inplace=True)
train2.set_index('PassengerId',inplace=True)
```

```python
print(train1.index)
print(train2.index)
```

```python
#merge using the index :
merged_by_index=pd.merge(train1,train2,left_index=True,right_index=True)
#use left_on or right_on if the index column name is different 
```

```python
print(merged_by_index.head(2))
```

```python
#four types of joins :
#inner , outer, left and right : 
merged_dataset_left=pd.merge(train1,train2,left_index=True,right_index=True,how="left")
```
